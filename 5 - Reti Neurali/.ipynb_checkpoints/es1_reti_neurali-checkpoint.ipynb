Riesci a comprendere il formato di questi file? Ma certo! Dopo averlo fatto, utilizzali per addestrare un multilayer perceptron, questo modello era effettivamente la scelta migliore?

```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
```

```python
#Import dei dati... Sarà tipo csv?? Bhoo proviamo
URL_TRAIN = "https://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra"
URL_TEST = "https://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tes"
data_train = pd.read_csv(URL_TRAIN, header=None)

data_train
```

FUNZIONAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAA

```python
data_test = pd.read_csv(URL_TEST, header=None)
data_test
```

Dalla descrizione del dataset, immagino che ogni riga sia ottenuta come concatenazione orizzontale delle righe della matrixe 8x8 di cui si parla nel sito.

Infatti 8x8 = 64 pixel in totale e qui ho 64 features (e poi una colonna target).

In ogni caso, per ora non ci importa molto, faccio l'allenamento del classificatore come al solito...

```python
X_train = data_train.drop(64,axis=1)
y_train = data_train[64]

X_test = data_test.drop(64,axis=1)
y_test = data_test[64]
```

Non so se fare esplorazione dei dati qui abbia senso, ma comunque do i dati per buoni, anche per quanto riguarda il bilanciamento delle classi.

Per curiosità visualizzo qualche immagine con imshow (eheh ci incontriamo di nuovo)

```python
import matplotlib.pyplot as plt

#visualizzo due immagini a caso
train_index = 176
plt.imshow(X_train.values[train_index,:].reshape(8,8),cmap='gray')
plt.colorbar()
plt.title(f"Img train n. {train_index} - real value: {y_train[train_index]}")
plt.show()

test_index = 176
plt.imshow(X_test.values[test_index,:].reshape(8,8),cmap='gray')
plt.colorbar()
plt.title(f"Img test n. {test_index} - real value: {y_test[test_index]}")
plt.show()
```

### Preprocessing
Il preprocessing da fare qui è una normalizzazione rispetto al valore massimo previsto. Avendo a che fare con immagini codificate su due bit, il massimo è 16, quindi divido ogni elemento delle matrici per questo valore.

Questo è diverso da fare min-max scaling, perchè nel min max scaling posso avere un massimo diverso per ogni feature (qui un massimo diverso per lo stesso pixel nelle varie immagini) e questo porta a distorcere le immagini: se, per esempio, il pixel in alto a destra ha un massimo di 10, finisce che questo valore avrà valore massimo = 1 e quindi avrebbe lo stesso valore finale di un pixel che invece ha valore massimo 16.
In sostanza, vado a ridurre il contrasto, e questo non va bene per la normalizzazione (che deve lasciare le relazioni tra i pixel per come sono).

```python
#No, leggi commento qui sopra
#from sklearn.preprocessing import MinMaxScaler

#mm = MinMaxScaler()
#X_train = mm.fit_transform(X_train) #trasforma in array numpy in automatico
#X_test = mm.transform(X_test) #trasforma in array numpy in automatico

#Conviene prima passare ad array numpy
X_train = X_train.values
y_train = y_train.values
X_test = X_test.values
y_test = y_test.values


X_max = X_train.max() #probabilmente sarà 16, ma non è detto
X_train = X_train / X_max
X_test = X_test / X_max 

#in questo modo le immagini non vengono distorte
```

```python
X_max
```

### Modello
Come modello qui è richiesto di usare il multi layer perceptron, ma... Con quali iperparametri? Quanti nodi? Quanti strati nascosti? 
Quale funzione di attivazione?
Bhooooo, per ora provo con le cose standard

```python
RANDOM_SEED = 200
#per lo strato nascosto il default è 1 strato nascosto con 100 neuroni...
#Non avendo idea di quale struttura sia migliore, provo una rete con 
# 2 strati nascosti ognuno con 300 neuroni, seguiti da uno strato
#con 50 neuroni (tropo? troppo poco?).

#NON ABBIAMO IDEA DI COSA STIAMO FACENDO, MA LO STIAMO FACENDO BENE
hidden_layer = (300, 300, 50)

#La funzione di attivazione di default (per gli strati nascosti) è la
#ReLU... Ma sì, lasciamo il mondo com'è
model = MLPClassifier(hidden_layer_sizes=hidden_layer, 
                      random_state=RANDOM_SEED,
                     verbose=True)

model.fit(X_train,y_train)
```

## Valutazione del modello
Uso il classification_report di sklearn

```python
from sklearn.metrics import classification_report
y_pred_train = model.predict(X_train)

print("Train set")
print(classification_report(y_train,y_pred_train))
```

```python
y_pred_test = model.predict(X_test)

print("Test set")
print(classification_report(y_test,y_pred_test))
```
