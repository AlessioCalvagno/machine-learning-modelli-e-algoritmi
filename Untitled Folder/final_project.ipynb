### Feature distributions
Here some charts are used to have a rough idea about features distribution, in order to discover some information useful for prediction problem.

For qualitative features the relative frequency distribution in shown, for quantitative ones, some statistical properties (mean, standard deviation and quartiles) are provided along with a plot of feature's distribution.
If the feature is discrete the distribution is shown as an histogram + boxplot, else a violinplot is provided.

```python
#define a custom function to make relative frequency distributions for 
#qualitative features. In charts, bins are frequency sorted

from matplotlib.ticker import PercentFormatter

#some graphics settings
plt.rcParams["figure.figsize"] = (16,10)
#plt.rcParams["font.size"] = 50
sns.set_theme(style='dark')

def plot_frequency_distribution(X, feature_name=None):
    """
    Make a bar chart of relative frequency distribution for a qualitative
    feature.
    The relative frequency values are printed in console as percentages.
    Prameters:
    - X : array-like or pandas.Series object
        Feature values.
    - feature_name: str or None
        Feature name, used for figure title.
        If None, no title is displayed. 
        Default None.
    """
    FONT_SIZE = 16
    
    relative_frequency = X.value_counts(normalize = True)*100
    #relative_frequency.sort_index(inplace=True)
    print(relative_frequency)
    ax = relative_frequency.plot.bar(rot=45, fontsize=FONT_SIZE)
    pf = PercentFormatter(xmax=100,decimals=2)
    ax.yaxis.set_major_formatter(pf)
    ax.bar_label(ax.containers[0], fmt='%.2f%%', fontsize=FONT_SIZE)
    plt.ylabel('Frequency (%)', fontsize=FONT_SIZE)
    plt.xlabel('Categories',fontsize=FONT_SIZE)
    plt.ylim((0,100))
    if feature_name != None:
        plt.title(f"Relative frequency distribution for {feature_name} feature",
                 fontsize=FONT_SIZE)
    plt.show()
    
 
```

```python
def boxplot_jitter(X, ax=None):
    """
    An helper function, to draw an horizontal boxplot and a stripplot to 
    better visualize otuliers.
    Stripplot is to be used only to have a raw idea about numerosity,
    if a more precise count is needed, use the return of this function.
    Parameters:
    - X : array-like or pandas.Series object
        Feature values.
    - ax: matplotlib Axes
        Axes object to draw the plot onto, 
        otherwise uses the current Axes. 
        Default None.
    Returns:
    - outliers: array-like or pandas.Series object
        Outliers computer with IQR method
    """
    sns.boxplot(x=X,width=.5,ax = ax)
        
    #overimpose a stripplot to better visualize outliers
    q1 = X.quantile(0.25)
    q3 = X.quantile(0.75)
    IQR = q3 - q1
        
    outlier_condition = (X > (q3 + 1.5 * IQR)) | (X < (q1 - 1.5 * IQR))
        
    # Add jitter only for outliers
    outliers = X[outlier_condition == True]
    sns.stripplot(x=outliers,
                    jitter=True,
                    alpha=0.5,
                    edgecolor="red",
                    facecolors="none",
                    linewidth=1.4,
                    ax=ax)
    return outliers
```

```python
def plot_quantitative_distribution(X, feature_name=None, discrete=False):
    """
    Make plot for distribution of a quantiative feature.
    If the feature is discrete, a pair of histogram and boxplot is shown,
    otherwise a violinplot is shown.
    In discrete case, a stripplot is overimposed to better visualize
    outliers numerosity. 
     Prameters:
    - X : array-like or pandas.Series object
        Feature values.
    - feature_name: str or None
        Feature name, used for figure title.
        If None, no title is displayed. 
        Default None.
    - discrete: bool
        If true, X is treated as a discrete feature (histogram + boxplot),
        else X is treared as continous (violinplot).
        Default False
    """
    FONT_SIZE = 16  
    if(discrete):
        #histogram
        fig, axs = plt.subplots(2,1,sharex=True)
        sns.histplot(x=X, discrete=True, ax = axs[0])
        axs[0].set_ylabel("Count", fontsize=FONT_SIZE)
        axs[0].tick_params(axis='y', labelsize=FONT_SIZE)
        
        #boxplot
        boxplot_jitter(X,axs[1])        
        
        plt.tight_layout()
    else:
        sns.violinplot(x=X)
        sns.despine(left=True)
        
    plt.xticks(fontsize=FONT_SIZE)
    if feature_name != None:
        plt.xlabel(feature_name, fontsize=FONT_SIZE)
        plt.suptitle(f"{feature_name} distribution", fontsize=FONT_SIZE)
    plt.show()
```

### OCCUPATION_TYPE cleaning
As expected most frequent value is when name_income_type is "working" and occupation_type is not missing. 

Regarding the pensioner case, one can see how all pensioners have an occupation_type value missing (there is no "Pensioner" bin in "Not missing" group); this is shown in the above print too. 

However one cannot infer that all missing values in OCCUPATION_TYPE are due to a pensioner state, since there are many records with OCCUPATION_TYPE not equal to "Pensioner" that have a missing value in OCCUPATION_TYPE (i.e. there isn't only the "Pensioner" bin in "Missing" group).  


#### How to handle this?

From this chart, one can say that all pensioners have a missing value regading OCCUPATION_TYPE, thus we can assign the new category "Pensioner"  in OCCUPATION_TYPE for these records.

As for the remaining values, probably their occupation isn't in the list of OCCUPATION_TYPE values, so we can aggregate these records under a new category named "Other".

```python
#remove temporary column
data.drop("OCCUPATION_TYPE_TMP",axis=1,inplace=True)
```

```python
data["OCCUPATION_TYPE_CLEAN"] = data["OCCUPATION_TYPE"]
data.loc[data["NAME_INCOME_TYPE"]=='Pensioner',"OCCUPATION_TYPE_CLEAN"]="Pensioner"

data["OCCUPATION_TYPE_CLEAN"]=data["OCCUPATION_TYPE_CLEAN"].fillna("Other")
#check
print(f'Remaining missing values: {data["OCCUPATION_TYPE_CLEAN"].isna().sum()}')
```

Let's check the new feature distribution.

```python
plot_frequency_distribution(data["OCCUPATION_TYPE_CLEAN"],"OCCUPATION_TYPE_CLEAN")
```

One last cleaning of this feature is including all rare categories inside the "Other" category: this will be useful for model construction and training; in fact, the categorical variables will be converted in numeric values with a OneHot encoding, and if one keeps all these categories, a huge amount of columns will be generated, and this could lead to computational problems.

A category is considered rare if its relative frequency is below 3%.

```python
relative_freq = data["OCCUPATION_TYPE_CLEAN"].value_counts(normalize=True)
rare_categories = relative_freq[relative_freq < 0.03].index

data["OCCUPATION_TYPE_CLEAN"]=data["OCCUPATION_TYPE_CLEAN"].apply(lambda x: "Other" if x in rare_categories else x)
#check the final feature
plot_frequency_distribution(data["OCCUPATION_TYPE_CLEAN"],"OCCUPATION_TYPE_CLEAN")
```

```python
from scipy.stats import chi2_contingency

def chi_test_feature_target(feature, target, categorical=False, nbins=10):
    """
    Compute a cross table between target and feature, and print 
    chi-suqared test results from this table.
    If the feature is quantitative, first it's segmented into bins.
    Parameters:
    - feature: pd.Series or array-like object.
        An object containing feature values
    - target: pd.Series or array-like object.
        An object containing target values.
    - categorical: bool.
        True if the feature is categorical, False otherwise.
        Default False.
    - nbins: int.
        Number of bins in which divide the feature. 
        If categorical is True, this parameter is ignored.
        Default 10.        
    """
    if(categorical):
        feature_discr = feature
    else:
        feature_discr = pd.cut(feature,nbins)
    cross_table = pd.crosstab(feature_discr, target)
    #Only pvalue is printed. To see full result remove .pvalue from line below.
    print(chi2_contingency(cross_table).pvalue)
```

```python
cols =  [col for col in data_clean.columns if col != 'TARGET']
quantitative_features=["CNT_CHILDREN",
                      "AMT_INCOME_TOTAL",
                      "CNT_FAM_MEMBERS",
                      "AGE",
                      "YEARS_EMPLOYED"]

qualitative_features = [column for column in cols if column not in quantitative_features]

#Quantitative features are discretized in 10 bins.
for col in cols:
    print(col)
    if col in quantitative_features:
        chi_test_feature_target(data_clean[col],data_clean["TARGET"],False)
    else:
        chi_test_feature_target(data_clean[col],data_clean["TARGET"],True)
            
```

Chi squared tests show how all pre-selected features have a non banal relationship with target: each test lead to a $p_{value} < 5\%$.

For this reason all these features are selected for prediction.

Don't get confused about using OrdinalEncoder for doing a Nominal Encoding:

according to sklearn documentation [OrdinalEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OrdinalEncoder.html#sklearn.preprocessing.OrdinalEncoder) is used for encoding features, while [LabelEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html#sklearn.preprocessing.LabelEncoder) is used for encoding target, but both perform a Label Encoding.

Moreover according to [this question in stackExchange](https://datascience.stackexchange.com/questions/39317/difference-between-ordinalencoder-and-labelencoder) sklearn doesn't provide an implementation for Ordinal Encoding, thus this must be performed manually with map() Series method: for this reason NAME_EDUCATION_TYPE is preprocessed with a mapping and not with OrdinalEncoder.

## Prediction model
Given that one wants to explain target result, as prediction model here a naive bayesian classifier is used. 

However considering that here there are different features types (i.e. binary, categorical and numeric), we'll use an hybrid bayesian classifier: 
- For binary features probability is computed according to Bernoulli distribution
- For categorical nominal features probability is computed according to Categorical distribution
- For all remaining features a normal distribution is used. 

This approach is valid since with naive bayesian classifier there is an assumption about features independency, and so they could even follow different distributions (like in this case).

<br></br>
### Note about features preprocsssing
Usually <ins>categorical features</ins> are preprocessed using a OneHotEncoder, in order to tell to the model that numeric values of these featues aren't a quantitative measure but are only a placeholder for a given category; in our case this step is unenecessary since we're using a Categorical distribution for these features, and the corresponding bayesian classifier already takes into account this values meaning.


So categorical features will be encoded only using a standard Label encoding, where each category (i.e. string in python world) is mapped to a positive integer number. 

<br></br>
Regarding <ins>ordinal qualitative features</ins>, these are preprocessed with an Ordinal Encoding, that is like the Label encoding but numeric values are given according to categories ordering. 
By doing so, these features become equivalent to numeric features, since their numeric values have an ordering meaning.

Considering this, the ordinal features will be preprocessed like numeric ones (after ordinal encoding). 

<br></br>
<ins>Numeric features</ins> are preprocessed with a standardization, since they will be used inside a Gaussian Bayesian classifier, where standardization is strongly suggested.

The preprocessing is implemented in HybridNaiveBayesClassifier custom class, except for ordinal encoding.

PS: In this dataset there is only one ordinal qualitative feature, that is _NAME_EDUCATION_TYPE_.

```python
#apply ordinal encoding
education_mapping = {
    "Lower secondary":0,
    "Secondary / se":1,
    "Secondary / secondary special":2,
    "Incomplete higher":3,
    "Higher education":4,
    "Academic degree":5    
}

#Ordinal Encoding
data_clean["NAME_EDUCATION_TYPE"] = data_clean["NAME_EDUCATION_TYPE"].map(education_mapping)

#List features names, divided by type (adapt previous logic)
cols = [col for col in data_clean.columns if col !='TARGET']

quantitative_features_clean=["CNT_CHILDREN",
                      "AMT_INCOME_TOTAL",
                      "CNT_FAM_MEMBERS",
                      "AGE",
                      "YEARS_EMPLOYED",
                      "NAME_EDUCATION_TYPE"]

binary_features_clean = [column for column in cols if column.startswith("FLAG_")]
binary_features_clean.append("CODE_GENDER")

categorical_features_clean = [column for column in cols if column not in quantitative_features_clean + binary_features_clean]

#check if everything is ok
print("Quantitative features")
print(quantitative_features_clean)
print("\nBinary features")
print(binary_features_clean)
print("\nCategorical features")
print(categorical_features_clean)

#should print 16
print(f"\n\nN. preprocessed features: {len(quantitative_features_clean)+len(binary_features_clean)+len(categorical_features_clean)}") 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from HybridNaiveBayesClassifier import *

X = data_clean.drop(columns=['TARGET'])
y = data_clean['TARGET']

RANDOM_SEED = 176

#Use stratify argument to keep classes proportion in train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                    stratify=y,
                                                    random_state=RANDOM_SEED)

model = HybridNaiveBayesClassifier(quantitative_features=quantitative_features_clean,
                                   binary_features=binary_features_clean,
                                  nominal_features=categorical_features_clean)

model.fit(X_train,y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

print("Train set")
print(classification_report(y_train,y_train_pred))
print("\nTest set")
print(classification_report(y_test,y_test_pred))
```

# Results comment
From these results there is a good news and a bad news:
- Good one: since performances are similar between train and test set, there isn't overfitting;
- Bad one: although accuracy is quite good ($91\%$ in both sets),  the dataset imbalance is propagated into the model; one can note it by looking at recall for positive class (i.e. sensitivity) that is much less than recall for negative class (i.e. specificity). This is due to dataset imbalance towards negative class (see previous data exploration). 

### How to handle it?
One way to handle it, is to use a different training set, that is the original training set, but with additional synthetic data: we can compute them with SMOTE technique, and train again the model an see what we get.

We use a SMOTE variant, since in the dataset there are numeric and categorical features.

This variant is called SMOTENC, and it's implemented in [imbalanced-learn library](https://imbalanced-learn.org/stable/over_sampling.html).

Check [here](https://imbalanced-learn.org/stable/references/generated/imblearn.over_sampling.SMOTENC.html#imblearn.over_sampling.SMOTENC) to see documentation. 

```python
from imblearn.over_sampling import SMOTENC

#Remember that 'NAME_EDUCATION_TYPE' is from a qualitative feature.
smote_nc = SMOTENC(categorical_features=binary_features_clean+
                   categorical_features_clean+
                  ['NAME_EDUCATION_TYPE'])
X_train_smote,y_train_smote = smote_nc.fit_resample(X_train,
                                           y_train)
```
